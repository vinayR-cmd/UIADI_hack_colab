```python
import pandas as pd
import re

def clean_state_name(x):
    if pd.isna(x):
        return x
    x = str(x).lower()
    x = re.sub(r'[^a-z\s]', ' ', x)   # remove symbols like &, .
    x = re.sub(r'\s+', ' ', x).strip()  # remove extra spaces
    return x
```

```python
# Create month column (safe even if month_name exists)
df['month'] = df['date'].dt.to_period('M')

monthly_df = (
    df.groupby(['state','district','month'], as_index=False)
    .agg({
        'age_0_5':'sum',
        'age_5_17':'sum',
        'age_18_greater':'sum',
        'total_enrollments':'sum'
    })
)
```

```python
monthly_df['child_enrolment_ratio'] = (
    monthly_df['age_0_5'] / monthly_df['total_enrollments']
)

monthly_df['teen_enrolment_ratio'] = (
    monthly_df['age_5_17'] / monthly_df['total_enrollments']
)

monthly_df['adult_enrolment_ratio'] = (
    monthly_df['age_18_greater'] / monthly_df['total_enrollments']
)
```

```python
monthly_df.head()
```

```python
monthly_df.shape
```

```python
# Risk score between 0 and 1
monthly_df['risk_score'] = (
    0.6 * monthly_df['child_enrolment_ratio'] +
    0.3 * monthly_df['teen_enrolment_ratio'] +
    0.1 * monthly_df['adult_enrolment_ratio']
)
```
